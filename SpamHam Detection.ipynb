## Import library and data set

```python
import pandas as pd
df = pd.read_csv("F:/Data_Science/Jupyter Content/IITG Content/spam.csv")
df.head()
```

```python
df.info()
```

```python
df.groupby('Category').describe()
```

```python
df['Spam']=df['Category'].map({'ham':0, 'spam':1})
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.Message,df.Spam, test_size=0.25)
```

## Vectorizing the Data

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
x_train_count = v.fit_transform(x_train.values)

x_train_count.toarray()[:3]
```

```python
## Using naive bayes
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_count,y_train)
```

### Example

```python
emails=[
    'Hey mohan, can we get together to watch football game tomorrow?',
    'Upto 20% discount on parking, exclusive offer jusst for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
x_test_count = v.transform(x_test)
model.score(x_test_count, y_test)
```

## Using pipeline

```python
from sklearn.pipeline import Pipeline
clf =Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())    
])
```

```python
clf.fit(x_train,y_train)
```

```python
clf.predict(emails)
```

```python
clf.score(x_test,y_test)
```
